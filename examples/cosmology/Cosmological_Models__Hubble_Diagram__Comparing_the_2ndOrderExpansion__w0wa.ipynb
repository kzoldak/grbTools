#### All luminosity distance functions must pass results in parsecs.

```python
def LumDist_concordance(redshift):
    """
    This is the function we use in our work, but different cosmo constants 
    as well as DL units. 
    """
    z           = redshift
    H_knot      = 71               # Hubble Constant.  Units: km/s/Mpc
    omega_m     = 0.27             # Matter Density.
    omega_l     = 1.0 - omega_m
    c           = 2.99792458e5      # SPEED OF LIGHT    Units: km/s
    def Aint(z):
        return (1./(sqrt(((1.+z)*(1.+z)*(1. + omega_m * z))-(z*(2.+ z)*omega_l ))))
    AA      = integrate.quad(Aint, 0.0, z)
    DL_Mpc  = (c * (1. + z) / H_knot) * AA[0]
    DL_pc = DL_Mpc * (1E6) # convert Mpc to pc
    return DL_pc


def LumDist_wexpansion_1(redshift):
    """
    Lower order expansion. Equation 14 in Riess et al. 2004. 
    w(z) = w0 + w'z
    """
    z           = redshift
    H_knot      = 71               # Hubble Constant.  Units: km/s/Mpc
    omega_m     = 0.27             # Matter Density.
    omega_l     = 1.0 - omega_m
    c           = 2.99792458e5      # SPEED OF LIGHT    Units: km/s
    w0 = -1.31
    wp = 1.48   # p: prime symbol
    def Aint(z):
        return 1./np.sqrt( ((1.+z)**3)*omega_m + \
                         omega_l*((1.+z)**(3*(1+w0-wp)))*np.exp(3*wp*z) )
    AA      = integrate.quad(Aint, 0.0, z)
    DL_Mpc  = (c * (1. + z) / H_knot) * AA[0]
    DL_pc = DL_Mpc * (1E6) # convert Mpc to pc
    return DL_pc


def LumDist_wexpansion_2(redshift):
    """
    THese w0 wa constants make the model look like Chaplyn gas.
    """
    H_knot      = 71               # Hubble Constant.  Units: km/s/Mpc
    omega_m     = 0.27             # Matter Density.
    omega_l     = 1.0 - omega_m
    c          = 2.99792458e5      # SPEED OF LIGHT    Units: km/s
    w0 = 0.1
    wa = -0.9

    z = redshift
    def Aint(z):
        return 1./np.sqrt( ((1.+z)**3)*omega_m + \
                         omega_l*((1.+z)**(3*(1+w0+wa)))*np.exp(-3*wa*z/(1+z)))
    AA      = integrate.quad(Aint, 0.0, z)
    DL_Mpc  = (c * (1. + z) / H_knot) * AA[0]
    DL_pc = DL_Mpc * (1E6)
    return DL_pc

def LumDist_wexpansion_3(redshift):
    """
    NOt the one published, but likely the correct w0 and wa constants. 
    """
    H_knot      = 71               # Hubble Constant.  Units: km/s/Mpc
    omega_m     = 0.27             # Matter Density.
    omega_l     = 1.0 - omega_m
    c          = 2.99792458e5      # SPEED OF LIGHT    Units: km/s
    # We think this are prob the correct ones.
    w0 = -0.9
    wa = 0.1

    z = redshift
    def Aint(z):
        return 1./np.sqrt( ((1.+z)**3)*omega_m + \
                         omega_l*((1.+z)**(3*(1+w0+wa)))*np.exp(-3*wa*z/(1+z)))
    AA      = integrate.quad(Aint, 0.0, z)
    DL_Mpc  = (c * (1. + z) / H_knot) * AA[0]
    DL_pc = DL_Mpc * (1E6)
    return DL_pc

def LumDist_wexpansion_4(redshift):
    """
    Schaefer 2007 states these w0 wa values are previous to his work.
    See eqn 43 and just above it. 
    """
    H_knot      = 71               # Hubble Constant.  Units: km/s/Mpc
    omega_m     = 0.27             # Matter Density.
    omega_l     = 1.0 - omega_m
    c          = 2.99792458e5      # SPEED OF LIGHT    Units: km/s
    w0 = 0.2
    wa = -1.4

    z = redshift
    def Aint(z):
        return 1./np.sqrt( ((1.+z)**3)*omega_m + \
                         omega_l*((1.+z)**(3*(1+w0+wa)))*np.exp(-3*wa*z/(1+z)))
    AA      = integrate.quad(Aint, 0.0, z)
    DL_Mpc  = (c * (1. + z) / H_knot) * AA[0]
    DL_pc = DL_Mpc * (1E6)
    return DL_pc

def LumDist_wexpansion_5(redshift):
    """
    Schaefer 2007 states these w0 wa values are previous to his work.
    See eqn 43 and just above it. 
    """
    H_knot      = 71               # Hubble Constant.  Units: km/s/Mpc
    omega_m     = 0.27             # Matter Density.
    omega_l     = 1.0 - omega_m
    c          = 2.99792458e5      # SPEED OF LIGHT    Units: km/s
    w0 = -1.4
    wa = 0.2

    z = redshift
    def Aint(z):
        return 1./np.sqrt( ((1.+z)**3)*omega_m + \
                         omega_l*((1.+z)**(3*(1+w0+wa)))*np.exp(-3*wa*z/(1+z)))
    AA      = integrate.quad(Aint, 0.0, z)
    DL_Mpc  = (c * (1. + z) / H_knot) * AA[0]
    DL_pc = DL_Mpc * (1E6)
    return DL_pc


def LumDist_weylgravity(redshift):
    """
    Weyl Gravity.
    Equation 237 in Mannheim 2006 paper
    q_knot      = -0.37 or -0.2
    """
    z           = redshift
    H_knot      = 71               # Hubble Constant.  Units: km/s/Mpc
    c           = 2.99792458e5      # SPEED OF LIGHT    Units: km/s
    q_knot      = -0.37
    DL_Mpc  = (-c * ((1. + z)**2) / (H_knot * q_knot)) * \
                (1-np.sqrt(1 + q_knot - (q_knot/((1+z)**2))))
    DL_pc = DL_Mpc * (1E6) # convert Mpc to pc
    return DL_pc

    
def distance_modulus(lumdistance):
    return 5*np.log10(lumdistance)-5

```
